### Bias-variance tradeoff
* Bias
* 
* Skillnad mellan vntevrdet av prediction och target(label)
___
* variance
* 
* 
* y_hatt = prediction frn ett trningsset.  sista epokens y_hatt.

### Bias-variance tradeoff
### Underfit
* Hg ^ bias
* Lg v variance
* Lgre komplexitet i modellen --> v mindre spridning
* 
 
### Overfitting
* Lg v Bias
* Hg ^ variance
* kar variansen --> minskar bias & vice versa
*
