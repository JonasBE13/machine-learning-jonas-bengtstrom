### Bias-variance tradeoff
* Bias
* 
* Skillnad mellan väntevärdet av prediction och target(label)
___
* variance
* 
* 
* y_hatt = prediction från ett träningsset.  sista epokens y_hatt.
* expected value of y_hatt = väntevärdet av många träningsset predictions

### Bias-variance tradeoff
### Underfit
* Hög ^ bias
* Låg v variance
* Lägre komplexitet i modellen --> v mindre spridning
* 
 
### Overfitting
* Låg v Bias
* Hög ^ variance
* Ökar variansen --> minskar bias & vice versa
* 

### error = bias + variance + irreducible error
* 
* irriducible error = kan ej påverka. kan bara påverka på insamling av datan. noise. 
* elbow method
* train|val|test- split
* 
* validera på valideringsdatan gfr att hitata hyperparametrar och därefter testar man på sin test data för att få dem riktiga metrics. om vi använder testdata som valideringsdata för att hitta polynom, då har vi läckt data (data leakage). Då har vi använt dne datan i vår träning. inte bra.
