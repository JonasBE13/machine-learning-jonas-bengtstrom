## Lasso (Least absolute shrinkage and selective operator)
* Svingar runt den, kastar den på våra beta parametrar. Dem vi fångar blir 0. eller tvärt om.
* Proportionelig mot L1-normen
* 
* När lambda är stor --> sätta minst viktiga (beta_i) features till 0
* sätter en del beta_i till 0
* bra när vi har många predictors, men endast få är viktiga.
* 
* bilexempel, en är lite bättre på att predicta, eller lite värden. Då är Lasso bra att använda

## K-fold cross-validation
### Annat sätt att göra Train|val|test split

* På detta sättet kan vi räkna ut vilka hyperparametrar som har gett oss lägsta medelvärdet.
* Minimierar risken att ha otur/tur (fungerar bra för denna hyperparametern, men inte generellt)
* blir mer robust. Kan utnyttja datan på bättre sätt.
* test-datan har vi lagt i vår skattkista. 
* för stora dataset kan det vara dyrt att använda denna metoden.
